# building linear regression from scratch

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_excel('headbrain.csv.xlsx')
cc=df.count()
cc
```

```python
x= df['Head Size(cm^3)'].values
y= df['Brain Weight(grams)'].values

```

```python
mean_x=np.mean(x)
mean_y=np.mean(y)
print(mean_x)
print(mean_y)
```

```python
beta_one=0
beta_zero=0
num=0
denom=0
for i in range(len(x)):
    num+=(x[i]-mean_x)*(y[i]-mean_y)
    denom+=(x[i]-mean_x)**2
    
beta_one= num/denom
beta_zero= mean_y - (beta_one*mean_x)
    
```

```python
beta_one
```

```python
beta_zero
```

```python
y_pred= beta_zero+beta_one*x
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred)
```

```python
rmse=0
for i in range(len(x)):
    rmse+=((y_pred[i]-y[i])**2)/len(x)
rmse=np.sqrt(rmse)
    
```

```python
rmse
```

# using SK Learn

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
lm = linear_model.LinearRegression()
```

```python
x_input= x.reshape(237,1)
y_input= y.reshape(237,1)
lm.fit(x_input,y_input)
y_pred = lm.predict(x_input)

print("Mean squared error: %.2f"% np.sqrt(mean_squared_error(y_input, y_pred)))
```
